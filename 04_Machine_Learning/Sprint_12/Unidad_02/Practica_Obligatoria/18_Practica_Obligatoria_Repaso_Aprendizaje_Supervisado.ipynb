### #1 Explicacin del dataset y carga de datos

Vamos a trabajar con el dataset "Wine Quality", que es un dataset en el que se recogen diversas caractersticas fisicoqumicas de las variades tinta y blanca del "Vinho verde" portugus. Las variables del dataset se describen a continuacin


1. **Acidez fija**: La mayora de los cidos involucrados con el vino son fijos o no voltiles (no se evaporan fcilmente).
2. **Acidez voltil**: La cantidad de cido actico en el vino, que en niveles demasiado altos puede llevar a un sabor desagradable a vinagre.
3. **cido ctrico**: Encontrado en pequeas cantidades, el cido ctrico puede aadir 'frescura' y sabor a los vinos.
4. **Azcar residual**: Es la cantidad de azcar restante despus de que se detiene la fermentacin, es raro encontrar vinos con menos de 1 gramo/litro y vinos con ms de 45 gramos/litro se consideran dulces.
5. **Cloruros**: La cantidad de sal en el vino.
6. **Dixido de azufre libre**: La porcin libre de SO2 existe en equilibrio entre el SO2 molecular (como un gas disuelto) y el ion bisulfito; previene el crecimiento microbiano y la oxidacin del vino.
7. **Dixido de azufre total**: Es la cantidad de formas libres y ligadas de S02; en bajas concentraciones, el SO2 es en su mayora indetectable en el vino, pero en concentraciones libres de ms de 50 ppm, el SO2 se hace evidente en la nariz y el sabor del vino.
8. **Densidad**: La densidad del vino es cercana a la del agua dependiendo del porcentaje de alcohol y contenido de azcar.
9. **pH**: Describe qu tan cido o bsico es un vino en una escala de 0 (muy cido) a 14 (muy bsico); la mayora de los vinos estn entre 3-4 en la escala de pH.
10. **Sulfatos**: Un aditivo del vino que puede contribuir a los niveles de dixido de azufre (SO2), que acta como un antimicrobiano y antioxidante.
11. **Alcohol**: El porcentaje de contenido de alcohol del vino.
12. **Calidad**: Una variable de salida (basada en datos sensoriales, puntuacin entre 0 y 10).
13. **Class**: Si es blanco o tinto.



Carga el dataset "wines_dataset.csv", muestra sus primeras filas, su informacin general y define dos variables target una para clasificacin que permita predecir la calidad del vino segn el resto de propiedas y otra variable target para regresin que permita predecir el grado alcohlico del vino a partir del resto de propiedades.

```python
# Dividir el data set en x e y
X = df.drop(columns=['quality', 'alcohol'])  
y_classification = df['quality']           

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train_class, y_test_class = train_test_split(
    X, y_classification, test_size=0.2, random_state=42, stratify=y_classification)

# Estandarizar los datos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train.select_dtypes(include=['float64']))
X_test_scaled = scaler.transform(X_test.select_dtypes(include=['float64']))

# Distribucin de la variable target para clasificacin: 'quality'
plt.figure(figsize=(8, 4))
sns.countplot(x=df['quality'])
plt.title("Distribucin de la variable 'quality'")
plt.xlabel("Calidad del vino")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Codificar la columna categrica 'class' utilizando OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, drop='first')
class_encoded = encoder.fit_transform(X[['class']])

# Reemplazar la columna original con las columnas codificadas
X_encoded = X.drop(columns=['class'])
X_encoded = np.hstack((X_encoded, class_encoded))
```

```python
# Volver a cargar el test y train
X_train, X_test, y_train_class, y_test_class = train_test_split(
    X_encoded, y_classification, test_size=0.2, random_state=42, stratify=y_classification)
# Modelos a evaluar
knn_3 = KNeighborsClassifier(n_neighbors=3)  # Baseline con KNN k=3
knn_5 = KNeighborsClassifier(n_neighbors=5)  # Baseline con KNN k=5
rf_model = RandomForestClassifier(random_state=42)  # Random Forest sin optimizar
svc_model = SVC(random_state=42)  # Support Vector Classifier

# Validacin cruzada para evaluar el recall ponderado de cada modelo
recall_knn_3 = cross_val_score(knn_3, X_train, y_train_class, cv=5, scoring='recall_weighted')
recall_knn_5 = cross_val_score(knn_5, X_train, y_train_class, cv=5, scoring='recall_weighted')
recall_rf = cross_val_score(rf_model, X_train, y_train_class, cv=5, scoring='recall_weighted')
recall_svc = cross_val_score(svc_model, X_train, y_train_class, cv=5, scoring='recall_weighted')

# Comparar los resultados
recall_results = {
    'KNN (k=3)': np.mean(recall_knn_3),
    'KNN (k=5)': np.mean(recall_knn_5),
    'Random Forest': np.mean(recall_rf),
    'SVC': np.mean(recall_svc)
}

recall_results
```

```python
# Definir el grid de hiperparmetros para Random Forest
param_grid_rf = {
    'n_estimators': [50, 100, 200],      # Nmero de rboles en el bosque
    'max_depth': [None, 10, 20],        # Profundidad mxima de los rboles
    'min_samples_split': [2, 5, 10]     # Mnimo de muestras para dividir un nodo
}

# Realizar bsqueda de hiperparmetros con GridSearchCV
grid_search_rf = GridSearchCV(
    RandomForestClassifier(random_state=42),
    param_grid_rf,
    cv=5,
    scoring='recall_weighted',
    n_jobs=-1
)

# Ajustar el modelo con los datos de entrenamiento
grid_search_rf.fit(X_train, y_train_class)

# Obtener los mejores parmetros y el mejor rendimiento
best_params_rf = grid_search_rf.best_params_
best_score_rf = grid_search_rf.best_score_

best_params_rf, best_score_rf
```

### #3 Modelado para regresin

Reutiliza todo lo que puedas la preparacin del modelado de la parte anterior y construye ahora el mejor modelo que puedas para predecir el grado alcohlico de un vino dadas sus caractersticas fisico qumicas, su clase y la puntuacin de calidad que le ha sido otorgada. Compara al menos tres tipos de modelos, seleccionando una mtrica adecuada de comparacin y escoge el ms apropiado empleando validacin cruzada antes o despus de optimizar hiperparmetros (en la sesin en vivo lo haremos antes por cuestiones de tiempo). Evala el modelo escogido y con los hiperparmetros ajustados de la mejor manera que puedas. Para terminar realizar un pequeo anlisis de errores.

```python
# Preparar los datos para regresin
X = df.drop(columns=['alcohol'])  # Variables predictoras
y = df['alcohol']  # Variable objetivo

# Codificar la variable categrica 'class'
class_encoded = encoder.fit_transform(X[['class']])

# Reemplazar la columna categrica con las columnas codificadas
X_encoded = X.drop(columns=['class'])
X_encoded = pd.concat([pd.DataFrame(X_encoded), pd.DataFrame(class_encoded)], axis=1)
# Dividir los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
# Modelos a comparar
models = {
    "Linear Regression": LinearRegression(),
    "Random Forest": RandomForestRegressor(random_state=42),
    "Support Vector Regressor": SVR()
}
# Asegurarse de que todas las columnas de X_encoded tengan nombres como cadenas (str)
X_encoded.columns = X_encoded.columns.astype(str)

# Volver a dividir los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Repetir la evaluacin de los modelos con validacin cruzada
cv_results = {}
for name, model in models.items():
    scores = cross_val_score(model, X_train, y_train, cv=5, scoring='r2')
    cv_results[name] = scores.mean()

# Seleccionar el mejor modelo basado en R2 promedio
best_model_name = max(cv_results, key=cv_results.get)
best_model = models[best_model_name]

# Entrenar el mejor modelo con los datos de entrenamiento
best_model.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = best_model.predict(X_test)

# Evaluar el modelo en el conjunto de prueba
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Resultados finales
{
    "Modelo seleccionado": best_model_name,
    "MSE (Error cuadrtico medio)": mse,
    "MAE (Error absoluto medio)": mae,
    "R2 (Coeficiente de determinacin)": r2,
    "Resultados validacin cruzada (R2)": cv_results}
```

```python

# Obtener las importancias de las caractersticas
feature_importances = best_model.feature_importances_
features = X_encoded.columns

# Ordenar las importancias de las caractersticas
sorted_indices = np.argsort(feature_importances)[::-1]
sorted_features = features[sorted_indices]
sorted_importances = feature_importances[sorted_indices]

# Visualizar las importancias de caractersticas
plt.figure(figsize=(10, 6))
plt.barh(sorted_features, sorted_importances, color="skyblue")
plt.gca().invert_yaxis()
plt.xlabel("Importancia", fontsize=12)
plt.ylabel("Caractersticas", fontsize=12)
plt.title("Importancia de caractersticas - Random Forest", fontsize=14)
plt.show()

# Anlisis de errores residuales: Comparar valores reales vs predichos
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.6, color="blue")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color="red", linestyle="--")
plt.xlabel("Valores Reales (alcohol)", fontsize=12)
plt.ylabel("Valores Predichos (alcohol)", fontsize=12)
plt.title("Valores Reales vs Predichos - Random Forest", fontsize=14)
plt.grid(alpha=0.3)
plt.show()
```

```python
# Definir el grid de hiperparmetros para Random Forest
param_grid_rf = {
    'n_estimators': [100, 200, 300],  # Nmero de rboles
    'max_depth': [10, 20, None],     # Profundidad mxima de los rboles
    'min_samples_split': [2, 5, 10]  # Mnimo de muestras para dividir un nodo
}

# Realizar la bsqueda de hiperparmetros con validacin cruzada
grid_search_rf = GridSearchCV(
    RandomForestRegressor(random_state=42),
    param_grid_rf,
    cv=5,
    scoring='r2',
    n_jobs=-1
)

# Ajustar el modelo con los datos de entrenamiento
grid_search_rf.fit(X_train, y_train)

# Obtener los mejores parmetros y el mejor rendimiento
best_params_rf = grid_search_rf.best_params_
best_score_rf = grid_search_rf.best_score_

# Entrenar el modelo con los mejores parmetros
optimized_rf = RandomForestRegressor(**best_params_rf, random_state=42)
optimized_rf.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred_optimized = optimized_rf.predict(X_test)

# Evaluar el modelo optimizado
mse_optimized = mean_squared_error(y_test, y_pred_optimized)
mae_optimized = mean_absolute_error(y_test, y_pred_optimized)
r2_optimized = r2_score(y_test, y_pred_optimized)

# Resultados finales
{
    "Mejores parmetros": best_params_rf,
    "MSE (Error cuadrtico medio)": mse_optimized,
    "MAE (Error absoluto medio)": mae_optimized,
    "R2 (Coeficiente de determinacin)": r2_optimized,
    "Mejor R2 (Validacin cruzada)": best_score_rf
}
```

```python

# Importancia de caractersticas del modelo optimizado
feature_importances_optimized = optimized_rf.feature_importances_
sorted_indices_optimized = np.argsort(feature_importances_optimized)[::-1]
sorted_features_optimized = X_encoded.columns[sorted_indices_optimized]
sorted_importances_optimized = feature_importances_optimized[sorted_indices_optimized]

# Visualizacin: Importancia de caractersticas
plt.figure(figsize=(10, 6))
plt.barh(sorted_features_optimized, sorted_importances_optimized, color="skyblue")
plt.gca().invert_yaxis()
plt.xlabel("Importancia", fontsize=12)
plt.ylabel("Caractersticas", fontsize=12)
plt.title("Importancia de caractersticas - Random Forest Optimizado", fontsize=14)
plt.show()

# Visualizacin: Valores reales vs predichos
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred_optimized, alpha=0.6, color="blue")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color="red", linestyle="--")
plt.xlabel("Valores Reales (alcohol)", fontsize=12)
plt.ylabel("Valores Predichos (alcohol)", fontsize=12)
plt.title("Valores Reales vs Predichos - Random Forest Optimizado", fontsize=14)
plt.grid(alpha=0.3)
plt.show()

# Anlisis de los errores residuales
residuals = y_test - y_pred_optimized

plt.figure(figsize=(8, 6))
plt.hist(residuals, bins=30, color="skyblue", edgecolor="black", alpha=0.7)
plt.xlabel("Errores Residuales", fontsize=12)
plt.ylabel("Frecuencia", fontsize=12)
plt.title("Distribucin de Errores Residuales - Random Forest Optimizado", fontsize=14)
plt.grid(alpha=0.3)
plt.show()
```
