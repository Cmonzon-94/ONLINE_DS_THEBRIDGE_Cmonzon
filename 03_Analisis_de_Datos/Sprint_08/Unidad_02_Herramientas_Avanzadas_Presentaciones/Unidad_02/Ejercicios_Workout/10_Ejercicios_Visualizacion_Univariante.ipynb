```python
df_titanic.head()
columnas_categoricas = ["class", "sex","who", "embark_town"]

num_columnas = 2
num_filas = (len(columnas_categoricas) + 1) // num_columnas

# Crear una figura con subplots
fig, axes = plt.subplots(num_filas, num_columnas, figsize=(12, num_filas * 4))
axes = axes.flatten()  # Para facilitar el acceso a los ejes

for i, columna in enumerate(columnas_categoricas):
    # Dibujar la grfica de barras en el eje correspondiente
    sns.countplot(data=df_titanic, x=columna, ax=axes[i])
    axes[i].set_title(f"Frecuencias Absolutas de {columna}", fontsize=12)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Frecuencia')

# Eliminar cualquier eje sobrante si las columnas no son mltiplo de 2
    for j in range(len(columnas_categoricas), len(axes)):
        fig.delaxes(axes[j])

plt.tight_layout()  # Ajustar el layout para evitar superposicin
plt.show()
```

### Ejercicio 4

Crea una funcin similar a la del ejercicio 4, pero que esta vez pinte el histograma de una lista de columnas numricas, aade un argumento para poder indicar el nmero de bins. Aade tambin un argumento que permita incluir la funcin de densidad o no en la figura. De nuevo usa Seaborn para crear las figuras. Tambin debe cumplirse el que pinte cada grafica en una matriz de 2 columnas como mucho y todas las filas que necesites.

```python

df_titanic.head()
columnas_numericas = ["open", "high","low", "close"]

num_columnas = 2
num_filas = (len(columnas_numericas) + 1) // num_columnas

# Crear una figura con subplots
fig, axes = plt.subplots(num_filas, num_columnas, figsize=(12, num_filas * 4))
axes = axes.flatten()  # Para facilitar el acceso a los ejes

for i, columna in enumerate(columnas_numericas):
    # Dibujar la grfica de barras en el eje correspondiente
    sns.histplot(data=df_titanic,kde=False,bins=100, ax=axes[i])
    axes[i].set_title(f"Frecuencias Absolutas de {columna}", fontsize=12)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Frecuencia')

# Eliminar cualquier eje sobrante si las columnas no son mltiplo de 2
    for j in range(len(columnas_numericas), len(axes)):
        fig.delaxes(axes[j])

plt.tight_layout()  # Ajustar el layout para evitar superposicin
plt.show()
```

### Ejercicio 5

Aplica la funcin del ejercicio anterior para pintar los histogramas de las columnas "ingresos", "consumo_kg" y "duracion" del dataset de viajes pero aplicando la siguiente lista de bins [10,30,60]. Aunque no seamos muy pro-histrogramas, con cul las posibles grficas te quedaras para visualizar los valores de cada variable?

```python

df_viajes.head()
columnas_numericas = ["ingresos", "consumo_kg","duracion","ingresos"]
bins = (10,30,60)

num_columnas = 2
num_filas = (len(columnas_numericas) + 1) // num_columnas

# Crear una figura con subplots
fig, axes = plt.subplots(num_filas, num_columnas, figsize=(12, num_filas * 4))
axes = axes.flatten()  # Para facilitar el acceso a los ejes

for i, columna in enumerate(columnas_numericas):
    # Dibujar la grfica de barras en el eje correspondiente
    sns.histplot(data=df_viajes,kde=False,bins=bins, ax=axes[i], color='skyblue', edgecolor='black')
    axes[i].set_title(f"Frecuencias Absolutas de {columna}", fontsize=12)
    axes[i].set_xlabel(columna)
    axes[i].set_ylabel('Frecuencia')

# Eliminar cualquier eje sobrante si las columnas no son mltiplo de 2
    for j in range(len(columnas_numericas), len(axes)):
        fig.delaxes(axes[j])

plt.tight_layout()  # Ajustar el layout para evitar superposicin
plt.show()
```
