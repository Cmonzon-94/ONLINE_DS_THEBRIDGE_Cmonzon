### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)


```python
#Buecamos la poblacion 
df_cities.loc[df_cities.population_total.nlargest(5).index]
```

```python
#Hallar las ciudades más pobladas para graficarlas 
ciudades_mas_pobladas = df_cities["city"].isin(["LosAngeles", "SanDiego", "Sacramento", "Anaheim", "SantaAna"])#viendo la leyenda del gráfico se saben las cities
df_pobladas = df_cities[ciudades_mas_pobladas].sort_values(by="population_total", ascending=False)
df_pobladas
```

```python
#Ahora podemos contruir el gráfico
categorias = df_pobladas.city.value_counts().index.to_list()
frecuencias = df_pobladas.population_total.to_list()
#Crear gráfico
fig = plt.figure(figsize=(8,5)) 
ax = plt.axes() 
barras = ax.bar(categorias, 
                frecuencias, 
                color=["blue", "grey", "red", "green", "brown"], 
                alpha=0.3)
# Personalización del gráfico
ax.set_title("Ciudades más pobladas de California")
ax.set_xlabel("Ciudades")
ax.set_ylabel("Habitantes (miles)")
#Agregamos los valores al gráfico
for barra, poblacion in zip(barras, frecuencias):
    ax.text(barra.get_x() + barra.get_width() / 2, barra.get_height(),
            f'{poblacion:,}', ha='center', va='bottom', fontsize=10)
```

2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)

```python
#Definimos el gráfico
plt.hist(df_cities.area_total_km2,
         bins= 50,
         alpha=0.7,
         color= "steelblue")
#Definimos formato leyenda
font_dict = {"size": 18,
             "family": "Bradley Hand itc",
}

plt.xlabel("Superficie (km2)",fontdict= font_dict )
plt.ylabel("Número de ciudades",fontdict= font_dict)
plt.legend(frameon=False, title="Hist Area");
```

3. El scatter de área de agua de un ciudad con su poblacion.

```python
#Graficamos el área de agua por su población
plt.figure(figsize=(10, 6))
plt.scatter(df_cities["area_water_km2"],
            df_cities["population_total"], 
            c="green",
            alpha=0.5, 
            s=df_cities["area_water_km2"])
plt.xlabel("Area de agua en la ciudad $(km^2)$")
plt.ylabel("Población(miles de habitantes)")

#Identificamos los outliers 
outliers = df_cities[df_cities["population_total"] > 800_000]
for _, outlier in outliers.iterrows():
    plt.annotate("Outlier", 
                 xy=(outlier["area_water_km2"], outlier["population_total"]),  
                 xytext=(outlier["area_water_km2"] + 10, outlier["population_total"] + 5000),
                 arrowprops=dict(facecolor='red', arrowstyle="->", lw=1.5),  # Estilo de la flecha
                 fontsize=10, color="red")
```

4. El boxplot de la altura de las ciudades.

```python
plt.boxplot(df_cities.elevation_m, 
            whis=1.5, 
            vert=False,
            patch_artist=True,
            boxprops=dict(facecolor="red"));
plt.title("Diagrama de caja de la Altitud de las ciudades de California")
plt.xlabel("Altitud (m)")
```

```python
#Visualización final
# Crear una figura con un arreglo de subtramas (2 filas x 2 columnas)
fig, axes = plt.subplots(2, 2, figsize=(15, 15))

# Gráfico 1: Barras
categorias = df_pobladas.city.value_counts().index.to_list()
frecuencias = df_pobladas.population_total.to_list()
barras = axes[0, 0].bar(categorias, frecuencias, color=["blue", "grey", "red", "green", "brown"], alpha=0.3)
axes[0, 0].set_title("Ciudades más pobladas de California")
axes[0, 0].set_xlabel("Ciudades")
axes[0, 0].set_ylabel("Habitantes (miles)")

# Agregar valores al gráfico de barras
for barra, poblacion in zip(barras, frecuencias):
    axes[0, 0].text(barra.get_x() + barra.get_width() / 2, barra.get_height(),
                    f'{poblacion:,}', ha='center', va='bottom', fontsize=10)

# Gráfico 2: Histograma
axes[0, 1].hist(df_cities.area_total_km2, bins=50, alpha=0.7, color="steelblue")
font_dict = {"size": 14, "family": "Bradley Hand ITC"}
axes[0, 1].set_xlabel("Superficie (km²)", fontdict=font_dict)
axes[0, 1].set_ylabel("Número de ciudades", fontdict=font_dict)
axes[0, 1].legend(frameon=False, title="Hist Area")

# Gráfico 3: Dispersión
scatter = axes[1, 0].scatter(df_cities["area_water_km2"], 
                              df_cities["population_total"], 
                              c="green", alpha=0.5, 
                              s=df_cities["area_water_km2"])
axes[1, 0].set_xlabel("Área de agua en la ciudad $(km^2)$")
axes[1, 0].set_ylabel("Población (miles de habitantes)")

# Identificar los outliers 
outliers = df_cities[df_cities["population_total"] > 800_000]
for _, outlier in outliers.iterrows():
    plt.annotate("Outlier", 
                 xy=(outlier["area_water_km2"], outlier["population_total"]),  
                 xytext=(outlier["area_water_km2"] + 10, outlier["population_total"] + 5000),
                 arrowprops=dict(facecolor='red', arrowstyle="->", lw=1.5),  # Estilo de la flecha
                 fontsize=10, color="red")

# Gráfico 4: Boxplot
axes[1, 1].boxplot(df_cities.elevation_m, 
                   whis=1.5, 
                   vert=False, 
                   patch_artist=True, 
                   boxprops=dict(facecolor="red"))
axes[1, 1].set_title("Diagrama de caja de la Altitud de las ciudades de California")
axes[1, 1].set_xlabel("Altitud (m)")

# Ajustar espacio entre subgráficos para evitar solapamientos
plt.tight_layout()

# Mostrar la figura final
plt.show()
```
